```python
import pandas as pd
import sqlite3
import pandas as pd
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.max_rows', None)     # Display all rows

conn = sqlite3.connect("/Users/m33210/Desktop/Anandas Documents/AP Projects/soccer_analytics/pl_database.db")
cursor = conn.cursor()



query = f'''SELECT a.Player, a.Team, a.Opponent, a.Pos, a.Gls, a.Sh, a.SoT, b.Date
    FROM summary a
    JOIN fixtures b ON a.match_id = b.match_id
    WHERE a.Team = 'Arsenal' and a.Opponent = 'Manchester City'
    group by a.Player
    ORDER BY b.Date;'''

df = pd.read_sql_query(query, conn)

df
```
