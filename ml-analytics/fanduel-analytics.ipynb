```python
try:
    qry = '''select "Player" as player, opponent, location, game_date, "Gls" as goals, "xG" as xg
    from summary;'''
    cur.execute(qry)
    qry_vals = cur.fetchall()
except Exception as e:
    conn.rollback()  # ðŸ‘ˆ This is crucial
    print(f"Error: {e}")
```

```python
df_json = {'player': [], 'opponent': [], 'location': [], 'game_date': [], 'goals': [], 'xg': []}
for game in qry_vals:
    df_json['player'].append(game[0])
    df_json['opponent'].append(game[1])
    df_json['location'].append(game[2])
    df_json['game_date'].append(game[3])
    df_json['goals'].append(game[4])
    df_json['xg'].append(game[5])
df = pd.DataFrame(df_json)
```
